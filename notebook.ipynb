## Exploratory Data Analysis (EDA)

### Explore Hubungan Weekday dan Banyaknya Penyewaan Sepeda Pada Tahun 2011-2012

```python
day_df.describe(include="all")
```

```python
day_df.groupby(by="weekday").agg({
    "casual": ["min", "max", "mean"],
    "registered": ["min", "max", "mean"],
    "cnt": ["min", "max", "mean"]
})
# 0 = Minggu, 1 = Senin, 2 = Selasa, 3 = Rabu, 4 = Kamis, 5 = Jumat, 6 = Sabtu
```

### Explore Hubungan Musim dan Lama Penyewaan Pada Tahun 2011-2012

```python
hour_df.describe(include="all")
```

```python
hour_df.groupby(by="season").agg({
    "hr": ["min", "max", "mean"]
})
# 1 = Springer, 2 = Summer, 3 = Fall, 4 = Winter
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
byweekday_df = day_df.groupby(by="weekday").agg({
    "casual": "mean",
    "registered": "mean",
    "cnt": "mean"
})

plt.figure(figsize=(10, 5))
plt.plot(byweekday_df.index, byweekday_df["casual"], label="Casual", marker='o',linewidth=2,)
plt.plot(byweekday_df.index, byweekday_df["registered"], label="Registered",  marker='o',linewidth=2,)
plt.plot(byweekday_df.index, byweekday_df["cnt"], label="Total",  marker='o',linewidth=2,)
plt.title("Number of Bicycle Rental Per Week in 2011-2012", loc="center", fontsize=15)
plt.xlabel("Weekday", fontsize=10)
plt.ylabel("Number of Bicycle Rental Per Week", fontsize=10)
plt.xticks(byweekday_df.index, ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'], fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.show()
```

### Pertanyaan 2:

```python
byseasons_df = hour_df.groupby(by="season").agg({
    "hr": "mean"
})
plt.figure(figsize=(10, 5))
sns.barplot(
    y="hr", 
    x="season",
    hue="season",
    data=byseasons_df.sort_values(by="hr", ascending=False),
    legend=False
)
plt.title("Number of Bicycle Rental Per Season in 2011-2012", loc="center", fontsize=15)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Average Hourly Rentals", fontsize=12)
plt.xticks([0,1,2,3], ['Spring', 'Summer', 'Fall', 'Winter'], fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

## Conclusion

- Terlihat bahwa banyaknya penyewaan selama 2011-2012 dipengaruhi oleh weekday, dengan total penyewaan terbanyak terjadi pada hari Jumat. Jika melihat dari karakteristek pelanggan, casual memiliki pelonjakan pada hari libur yaitu sabtu-minggu ini dikarenakan pengguna casual hanya menggunakan Bike Saring untuk rekreasi. Sedangkan, pada karakteristik regitered, penurunan terjadi di hari libur yaitu sabtu-minggu yang dapat disimpulkan pelanggan registered adalah pekerja yang bekerja pada senin-jumat.
- Terlihat bahwa lama penggunaan sepeda pada 4 season selama 2011-2012 memiliki rata-rata yang hampir sama dengan rata-rata tertinggi diraih oleh musim spring. Ini menunjukan pengaruh musim dalam lamanya penggunaan Bike Sharing sangatlah kecil yang menyebabkan tidak ada kekhawatiran pada musim-musim tertentu.
